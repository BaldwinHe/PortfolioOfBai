```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from itertools import cycle
palette = cycle(sns.color_palette("hls", 8))
```

```python
def get_x(y):
    return [i+1 for i in range(len(y))]

def draw_plot(y, label):
    color = next(palette)
    plt.plot(get_x(y),y, label=label, color=color)
    plt.axvline(x=len(get_x(y)),ls="--",color=color)
# plt.rcParams['savefig.dpi'] = 300 #
# plt.rcParams['figure.dpi'] = 300 #
plt.rcParams['figure.figsize'] = (8.0, 4.0) 
dqn = np.load('../checkpoints/Vanilla DQN/scores.npy')
double_dqn = np.load('../checkpoints/Double DQN/scores.npy')
with plt.style.context('seaborn'):
    draw_plot(dqn, 'Vanilla DQN')
    draw_plot(double_dqn, 'Double DQN')
    plt.title('Average score per episode')
    plt.xlabel('episode', fontsize=14)
    plt.ylabel('average score', fontsize=14)
    plt.legend()
```
