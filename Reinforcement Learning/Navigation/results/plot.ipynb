```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from itertools import cycle
from more_itertools import chunked
palette = cycle(sns.color_palette("hls", 8))
```

```python
def get_x(y):
    return [i+1 for i in range(len(y))]

def process(data, n=10):
    return [sum(x) / len(x) for x in chunked(data, n)]
def draw_plot(y, label):
    color = next(palette)
    plt.plot(get_x(y),y, label=label, color=color)
    plt.axvline(x=len(get_x(y)),ls="--",color=color)
# plt.rcParams['savefig.dpi'] = 300 #
# plt.rcParams['figure.dpi'] = 300 #
plt.rcParams['figure.figsize'] = (10.0, 5.0) 
dqn = np.load('../checkpoints/Vanilla DQN/scores.npy')
double_dqn = np.load('../checkpoints/Double DQN/scores.npy')
pri_dqn = np.load('../checkpoints/Prioritized Experience Replay DQN/scores.npy')
dqn = process(dqn)
double_dqn = process(double_dqn)
pri_dqn = process(pri_dqn)
with plt.style.context('seaborn'):
    draw_plot(dqn, 'Vanilla DQN')
    draw_plot(double_dqn, 'Double DQN')
    draw_plot(pri_dqn, 'Prioritized Experience Replay DQN')
    plt.title('Average score per 10 episodes')
    plt.xlabel('episodes(10)', fontsize=14)
    plt.ylabel('average score', fontsize=14)
    plt.legend()
```
